```python
from aegraph import AEGraph
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

gini_csv = pd.read_csv("API_SI.POV.GINI_DS2_en_csv_v2_35.csv")

countries_mean_gini = {}

for country in gini_csv["Country Name"]:

    country_df = gini_csv[gini_csv["Country Name"] == country]
    country_df = country_df.iloc[:, 4:].dropna(axis=1)

    data = country_df.iloc[0].astype(float).dropna()
    years = data.index.astype(float)

    # Only plot countries with sufficient data
    if len(data) > 5:
        countries_mean_gini[country] = np.mean(data[-5:])
        # countries_mean_gini[country] = data.iloc[-1]
        
# Example with gradient coloring (from red to green)
s = pd.Series(countries_mean_gini).sort_values(ascending=True).sort_values(ascending=False)

y_positions = np.arange(len(s))
widths = s.values
labels = s.index.tolist()
```

```python
color1 = [98+(255-98)/1.5, 157+(255-157)/1.5, 252+(255-252)/1.5] # rgb value of a light blue
color2 = [98+(255-98)/5, 157+(255-157)/5, 252+(255+252)/5] # rgb value of a slightly darker blue

(
    AEGraph(width=960, height=4000, compwidth=2160, compheight=3840,
            drop_shadow=True, comp_name="Gini_Barh_Gradient", fps=60,
            show_all_points=False,animate_opacity=False)
    .barh(y_positions, widths, animate=5.0, bar_duration=0.5, alpha=0.8, animate_downward = True)
    .gradient(color1, color2)  # Apply color gradient from high to low inequality
    .set_xlabel("Mean Gini Coefficient")
    .set_ylabel("Country")
    .set_title("Top 15 Countries by Mean Gini (Last 5 Years)")
    .set_xticks()
    # .set_xlim(20,45)       commented out because i dont want the graph to be misleading
    .set_yticks(positions=y_positions, labels=labels)
    .grid(show=True, color="gray", alpha=0.2)
    .reset_comp()
    .save()
    .render()
)
```
